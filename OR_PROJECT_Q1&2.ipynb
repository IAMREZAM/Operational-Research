```python
from pyomo.environ import *

model = ConcreteModel()

# Define the index set
model.I = RangeSet(1, 10)

# Define the decision variables
model.x = Var(model.I, domain=NonNegativeReals)

# Define the objective function
model.obj = Objective(
    expr=sum([2.5 * model.x[1] + 2.5 * model.x[2] + 1.6 * model.x[3] +
              2 * model.x[4] + 1.4 * model.x[5] + 1 * model.x[6] +
              0.8 * model.x[7] + model.x[8] + 1.4 * model.x[9] +
              0.8 * model.x[10]]),
    sense=minimize)

# Define the constraints
model.con1 = Constraint(expr=model.x[1] == 125)
model.con2 = Constraint(expr=model.x[2] == 175)
model.con3 = Constraint(expr=model.x[3] + model.x[4] == 225)
model.con4 = Constraint(expr=model.x[5] + model.x[6] == 250)
model.con5 = Constraint(expr=model.x[7] + model.x[8] == 225)
model.con6 = Constraint(expr=model.x[9] + model.x[10] == 200)
model.con7 = Constraint(expr=sum([model.x[i] for i in [2, 3, 5, 7, 9]]) <= 550)
model.con8 = Constraint(expr=sum([model.x[i] for i in [1, 4, 6, 8, 10]]) <= 700)

# Solve the model
solver = SolverFactory('glpk')
result = solver.solve(model)

# Display the solution
model.display()

# Print the objective value
print("Objective value:", value(model.obj))
```

```python
import numpy as np

# Define the variables
Q = np.array([[48, 12], [8, 8]])  # Matrix Q
q = np.array([[13], [23]])  # Vector q
p = 4  # Constant p
x_0 = np.array([[23], [37]])  # Initial guess for x

x_old = x_0

iteration = 1

# Calculate the gradient at x_old
gradient = np.dot(Q, x_old) + q
# Calculate alpha using the provided formula
alpha = np.dot(gradient.T, gradient) / np.dot(np.dot(gradient.T, Q), gradient)

print("alpha is : ",alpha)

while True:
    # Calculate the gradient at x_old
    gradient = np.dot(Q, x_old) + q

    # Update x_new using the gradient descent formula
    x_new = x_old - alpha * gradient

    # Calculate the difference between x_new and x_old
    diff = np.linalg.norm(x_new - x_old)

    # Print the current iteration's x, difference, and iteration number
    print("Iteration:", iteration)
    print("x =\n", x_new)
    print("Difference:", diff)
    print("-------------------")

    # Check if the difference is below the threshold
    if diff < 1e-6:
        break

    # Update x_old and iteration for the next iteration
    x_old = x_new
    iteration += 1

# Print the final updated value of x
print("Final updated value of x:")
print(x_new)
# Calculate the value of the objective function at x_new
f_x = 0.5 * np.dot(np.dot(x_new.T, Q), x_new) + np.dot(q.T, x_new) + p

# Print the value of the objective function at x_new
print("f(x_final):", f_x)
```
